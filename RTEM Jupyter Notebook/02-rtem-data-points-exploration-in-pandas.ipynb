## Querying data points

Maybe the most interesting function of this API is getting data from an specific point (i.e, a sensor). We need to import a specific class for doing this, the *PointSelector*

````python
from onboard.client.models import PointSelector
````

First create your *PointSelector* object, this will be the query:

````python
query = PointSelector()
````

And you can specify all the conditions you want ([check the doc](https://onboard-data-python-client-api.readthedocs.io/en/latest/Querying%20Building-Specific%20Data.html#querying-specific-points)). There are multiple ways to select points using the PointSelector. The user can select all the points that are associated with one or more lists containing any of the following:
```
'organizations', 'buildings', 'point_ids', 'point_names', 'point_hashes',
'point_ids', 'point_names', 'point_topics', 'equipment', 'equipment_types'
```
Under these categories you can get the points that meet the conditions you set.

````python
query.point_types = ['Real Power', 'Zone Temperature']
query.buildings = ["385"]
````

Here we are saying "I want the points of type 'Real Power' and 'Zone Temperature' from building 385". And when you execute that query using `select_points` you will get the points that satisfy those conditions:

````python
selection = client.select_points(query)
selection
````

Those are the points, identified by id:

````python
points = selection["points"]
points
````

And getting the points metadata is pretty easy when you have the IDs. You can get all the details about those points using `get_points_by_ids`:

````python
sensor_metadata = client.get_points_by_ids(points)
sensor_metadata = pd.DataFrame(sensor_metadata)
sensor_metadata
````

What if we want to add information about the equipments to our data? We can get a dataframe with all the equipment (like we did before):

````python
# Get a list of all the equipment in your building
equipment = pd.json_normalize(client.get_all_equipment())
equipment.head()
````

Don't forget to use the `.info()` method, is great to know your features and the data types.

We have the `id` for each equipment in the first columns; we also have it in our sensor metadata:

````python
sensor_metadata.head()
````

````python
equipment.info()
````

````python
sensor_metadata.info()
````

For merging two sets of data we can use [pandas merge](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html). In this case, we want to preserve the `sensor_metadata` rows and get the rows with those equip_ids from `equipment`. Notice that the column in common doesn't have the same name (is named "equip_id" on sensors data and "id" on equipment data. Yes. there is an "equip_id" on the equipment data but is a string and if you inspect it is more descriptive) so we have to specify it. Also, there are columns with the same name: in order to not being mixed up `merge` function will add the suffixes \_x and \_y (you can specify the suffixes you want).

````python
merged = pd.merge(sensor_metadata, equipment, how="left", left_on=["building_id","equip_id"], right_on=["building_id","id"])
merged.head()
````

````python
merged.info()
````

Always, when merging, make sure to identify the column in common and that they have the same data type.

# Data points exploration

## Data pre-processing

Now that we know how to get the data, let's obtain a big data set of points to explore it. In contrast to last section, here we are getting all the available points in our building to explore them.

````python
points = pd.json_normalize(client.get_all_points())
points.head()
````

````python
points.info()
````

And let's merge it with the buildings metadata.

````python
bdgs = pd.json_normalize(client.get_all_buildings())
bdgs
````

We are goint to add the building name to our points data set. You can select the desired columns using double brackets, like this:

````python
bdgs[["id", "name"]]
````

An to merge, like we just learnt, you have to specify in the parameter HOW you want to merge (we want to keep all the rows on the LEFT here, the `points` rows) and ON which column. We are also adding suffixes in order to correctly identify which one came from which dataset.

````python
points = pd.merge(points, bdgs[["id", "name"]], how="left", left_on="building_id", right_on="id", suffixes=["_point", "_bdg"])
points.info()
````

And now let's add some column from the equipment metadata:

````python
equipment.info()
````

Let's inspect a little these columns transposing our dataset (like we did at the beginning):

````python
equipment.head(1).T
````

And again, let's merge. Remember: column/s in common to merge ON them, HOW to merge and optional suffixes.

````python
points = pd.merge(points, equipment[["id","equip_type_abbr","equip_type_name", "area_served_desc", "tags"]], how="left", left_on="equip_id", right_on="id")
points.info()
````

````python
points.head(1).T
````

We got a duplicated column ("id" now is the equipment id) that we can [drop](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html):

````python
# Axis=1 means drop columns. Axis=0 means drop rows
points.drop("id", axis=1, inplace=True)
````

And let's inspect our missing values. If we have column with near 100% of missing values we are going to drop them too (this is not actually needed, but extra columns will make the dataset heavier and everything slower).

````python
points.isna().sum()*100 / len(points)
````

Let's drop those almost empty columns:

````python
points.drop(["network_device", "state_text", "state_text.1", "state_text.2", "state_text.3", "state_text.4", "state_text.5"], axis=1, inplace=True)
````

And let's export our data set as csv! we may need it later again:

````python
points.to_csv("rtem_all_points_metadata.csv", index=False)
````

## Data Visualization

Now let's create some visualizations! We are going to use the libraries *matplotlib* and *seaborn*:

````python
import matplotlib.pyplot as plt
import seaborn as sns

# This is for the visual stype, I like "ggplot"
plt.style.use('ggplot')
````

````python
points.head(1).T
````

The data set we are working with contains the metadata of each point (sensor). Is all categorical data, we have a lot of [vsualizations available](https://seaborn.pydata.org/tutorial/categorical.html) for that. Here all of our features are categorical, so we will focus mainly on the bar plots.

We can start exploring the number of points by buildings, we can check how many points and buildings we have first:

````python
print(f"Number of points: {len(points.id_point.unique())}")
print(f"Number of buildings: {len(points.building_id.unique())}")
````

Ok, that's a lot of points and some buildings. Let's plot a simple bar chart to visualize at a glanze how the points distribute among them (you can choose [the color you want](https://matplotlib.org/3.5.0/gallery/color/named_colors.html)):

````python
# Number of points by building

# Let's create a figure with custom size
fig, ax = plt.subplots(1,1, figsize=(30,6))

# The figure
g0 = sns.countplot(data=points, x="name_bdg", color="crimson", ax=ax)
ax.set_title("Number of points by building")
ax.set_xlabel("Building id")
ax.set_ylabel("# points")

# Rotate the labels to make it readable
labels = plt.xticks(rotation=90)
````

Kind of crowded with that many buildings, right? But is evident we have one with most of the points in our data. We have column named `description` that is, of course, a description of the sensor; this will give us an idea of the kind of sensors we are working with. Let's try to plot the top 50 descriptions (this is, the most frequent descriptions in our dataset). For that, we need to learn how to count the number of buildings in each category, [we can use the method `value_counts`](https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html):

````python
points.description.value_counts()
````

And to filter the top 50 (they are sorted from largest to smallest amount) we just select the first 50 elements and get the `index` (which is the name of the category) so we have a list:

````python
points.description.value_counts().iloc[:50].index
````

Let's plot it. We are going to use this list in the parameter `order`, pay attention:

````python
# Top 10 number of points by description

# Let's create a figure with custom size
fig, ax = plt.subplots(1,1, figsize=(12,30))

g1 = sns.countplot(data=points, y="description", order=points.description.value_counts().iloc[:50].index,ax=ax, color="crimson")
ax.set_title("Top50 point descriptions")
ax.set_ylabel("")
ax.set_xlabel("# points")
ax.legend(loc="lower right")

## Show percentage at end of the bar
for p in g1.patches:
    txt = str(round(p.get_width() / len(points) * 100, 2)) + '%'
    txt_x = p.get_width() 
    txt_y = p.get_y()
    g1.text(txt_x, txt_y, txt)
````

Now we are going to do something similar but with extra information. We are going to plot the number of points by `equip_type_name` and coloring by `name_bdg`; for the last, we have to use the parameter `hue`:

````python
# Number of points by equip type and building

# Let's create a figure with custom size
fig, ax = plt.subplots(1,1, figsize=(12,15))

g2 = sns.countplot(data=points, y="equip_type_name", hue="name_bdg", palette="hls", ax=ax)
ax.set_title("Points by equipment type")
ax.set_ylabel("")
ax.set_xlabel("# points")
ax.set_xlim(0,500)

# Put the legend out of the figure
legend = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., ncol=4)
````

Ok, pretty messy to visualize but it makes the point: you can visualize more than one feature at the same time.

Finally, let's create a count plot but not of point: let's count how the tags distribute. As you can see in the data, a point can have several tags. Let's find out which ones are the most frequent. We have to get a list with all of them:

````python
# This will create a list with all the tags in our data: it just concatenate the lists from each row
tags = points.tags.sum()

# check the first 10
tags[:10]
````

And we have to create a data frame with it:

````python
tags = pd.DataFrame(tags, columns=["tags"])
tags.head()
````

A pretty simple data set. Now we can make use of the `countplot` function. We are going to plot the top 20 only (as we did before):

````python
# Top 10 number of tags

# Let's create a figure with custom size
fig, ax = plt.subplots(1,1, figsize=(20,8))

g4 = sns.countplot(data=tags, y="tags", order=tags.tags.value_counts().iloc[:20].index, ax=ax, color="crimson")
ax.set_title("Top 20 tags")
ax.set_ylabel("")
ax.set_xlabel("# tags")

## Show percentage at end of the bar
for p in g4.patches:
    txt = str(round(p.get_width() / len(points) * 100, 2)) + '%'
    txt_x = p.get_width() 
    txt_y = p.get_y()
    g4.text(txt_x, txt_y, txt)
````

Most of the points are hvac for sure! Remember that each point can have several tags.

Now we know how to plot each chart individually, but what if we want only one figure with several subplots? Yes, there's a function for that! Just for visualizing, we are going to select only 5 buildings:
